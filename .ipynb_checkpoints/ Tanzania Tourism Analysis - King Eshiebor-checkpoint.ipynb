# DataLine Analytics Competition

## Context:

The Tanzanian tourism sector plays a significant role in the Tanzanian economy, contributing about 17% to the country’s GDP and 25% of all foreign exchange revenues. The sector, which provides direct employment for more than 600,000 people and up to 2 million people indirectly, generated approximately $2.4 billion in 2018 according to government statistics. Tanzania received a record 1.1 million international visitor arrivals in 2014, mostly from Europe, the US and Africa.
Tanzania is the only country in the world which has allocated more than 25% of its total area for wildlife, national parks, and protected areas.There are 16 national parks in Tanzania, 28 game reserves, 44 game-controlled areas, two marine parks and one conservation area.


## Objective:

The objective of this competition is to explore and build a linear regression model that will predict the spending behaivior of tourists visiting Tanzania.The model can be used by different tour operators and the Tanzania Tourism Board to automatically help tourists across the world estimate their expenditure before visiting Tanzania. 


## Data Description

The dataset describes 6476 rows of up-to-date information on tourist expenditure collected by the National Bureau of Statistics (NBS) in Tanzania.The dataset was collected to gain a better understanding of the status of the tourism sector and provide an instrument that will enable sector growth.
The survey covers seven departure points, namely: Julius Nyerere International Airport, Kilimanjaro International Airport, Abeid Amani Karume International Airport, and the Namanga, Tunduma, Mtukula and Manyovu border points.

The same applies to the test dataframe as the ratio of missing values is similar. Both dataframes will be combined and cleaned

```python
#combine two dataframes by appending without sorting
df = train.append(test, sort=False)
```

```python
def ratio(cond_col, cat, df=df):
    '''
    Takes a conditional column (cond_col) - 
    from a dataframe (df) - 
    to check for instances where a value in that column falls in a particular category(cat)
    '''
    #length of dataframe subset divided  by length of dataframe minus total number of missing values
    query = len(df[df[cond_col] == cat])/(len(df)-1441)
    return query
```

```python
def ratio2(cond_col, cat, df=df):
    '''
    Takes a conditional column (cond_col) - 
    from a dataframe (df) - 
    to check for instances where a value in that column falls in a particular category(cat)
    '''
    #length of dataframe subset divided  by length of dataframe (missing values have been filled)
    query = len(df[df[cond_col] == cat])/len(df)
    return query
```

```python
df.travel_with.unique()
```

```python
#Iterate through and print proportion of each category in travel_with
for i in ['Friends/Relatives', 'Alone', 'Spouse', 'Children', 'Spouse and Children']:
    print('Ratio of {} to entire travel_with is {}'.format(i, (ratio(cond_col='travel_with', cat=i)*100)))
```

Now to fill the missing values

```python
df_alt = df.copy()
df_alt['travel_with'] = df['travel_with'].ffill(inplace=False)
```

```python
df_alt['travel_with'].unique()
```

##### Final Cleaning and transformation

```python
# Converting the categorical variables into numerical -1 (changing yes and no to 1 and 0 respectively)
varlist =  ['package_transport_int', 'package_accomodation', 'package_food', 'package_transport_tz', 'package_sightseeing',
            'package_guided_tour', 'package_insurance', 'first_trip_tz']

# Defining the map function
def binary_map(x):
    '''Returns 1 if value is Yes and 0 is value is No'''
    return x.map({'Yes': 1, "No": 0})

# Applying the function to the housing list
df_new[varlist] = df_new[varlist].apply(binary_map)
```

```python
def dumm(df, varlist):
    '''Gets dummy variables for each column name in the list (varlist) from df (df)'''
    for i in varlist:
        status = pd.get_dummies(df[i], drop_first = True)
        df = pd.concat([df, status], axis = 1)
        df.drop([i], axis = 1, inplace = True)
    return df
```

```python
#create list of columns to change to dummy vaariables
to_dummy = ['age_group', 'travel_with', 'purpose', 'main_activity', 'tour_arrangement']
df_final = dumm(df_new, to_dummy)
```

```python
#Build a linear model
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.summary()
```

When we look at the p-values for several of the variables, we notice that they are rather high, indicating that they aren't significant. As a result, such variables may be removed from the model.

We must first examine the multicollinearity between the variables, as mentioned above, before removing them. This is accomplished by determining the VIF value. The Variance Inflation Factor, or VIF, is a numerical statistic that indicates how closely the feature variables are connected.
It's a crucial parameter for putting our linear model to the test.
The VIF formula is as follows:
$$
VIF_i = \frac{1}{1-R_i^2}
$$

```python
# Checking for the VIF values of the variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Creating a dataframe that will contain the names of all the feature variables and their VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Variables with a value of 5 and above are normally considered for removal. If we look closely at the graphic above, we can see that there are several variables that need to be removed. When it comes to removing variables, the p-value will take precedence. In addition, we must drop one variable at a time. Removing the variable from the model and updating it <br>
We repeat this until all p-values are less than 0.05 and all VIFs are less than 5

```python
# Dropping highly correlated variables and insignificant variables
X = X_train.drop('Hunting tourism', 1,)

# Build a fitted model after dropping the variable
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()

# Printing the summary of the model
print(lr_2.summary())
```

Now all p values are less than 5% and all VIFs are less than 5. Before making any predictions, we must first determine if the error terms are regularly distributed. We'll use Residual Analysis to do this. <br>
We have to check if the error terms are normally distributed (which is one of the major assumptions of linear regression); let us plot the error terms’ histogram.

```python
y_train_total_cost = lr_13.predict(X_train_lm)
# Plot the histogram of the error terms
fig = plt.figure()
plt.figure(figsize=(15,8))
sns.distplot((y_train - y_train_total_cost), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
# Plot heading and labels
plt.xlabel('Errors', fontsize = 18)
plt.ylabel('Frequency')
plt.title('Residual Errors Distribution');  
```

The error terms, as can be seen, closely match a normal distribution. As a result, we may use the model in the test dataset to generate predictions.
